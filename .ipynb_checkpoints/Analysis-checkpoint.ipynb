```python
round_1 = pd.read_csv('Code Rush.csv')
```

```python
columns = []
columns.append('HackerRank Username')
columns.append('Timestamp')
columns.append('Total score')
```

```python
for column in round_1.columns:
    if 'Score' in column:
        columns.append(column)
```

```python
columns
```

```python
del columns[3:5]
```

```python
columns
```

```python
round_1 = round_1[columns]
```

```python
round_1
```

```python
q = 1
col = []

for i in columns[3:]:
    round_1[f'Q_{q}'] = round_1[i].str[:1]
    col.append(f'Q_{q}')
    round_1[f'Q_{q}'].fillna(0,inplace=True)
    round_1[f'Q_{q}'].astype(int,inplace=True)
    q+=1
```

```python
df_1 = round_1.drop(columns[3:13],1)
```

```python
df_1.drop('Total score',1,inplace=True)
```

# Detailed analysis of scores in 1st round

```python
df_1
```

```python
df_1['Timestamp'] = pd.to_datetime(df_1['Timestamp']).dt.time
```

```python
df_1
```

```python
df_1.sort_values('score ', ascending =False)
```

```python
df_2 = pd.merge(df_1,df, on='HackerRank Username', how='outer')
```

```python
df_2.sort_values('score ',inplace=True)
```

```python
plt.barh(df_2['HackerRank Username'][:-1],df_2['score '][:-1])
plt.tight_layout()
plt.savefig('Round_1_scores')
```

```python
df_2['score ']= df_2['score '].astype(int,inplace=True)
```

```python
df_2.groupby('Year')['score '].mean()
```

# 2nd Year beats 1st year
## 2nd year teams scored an average of 6.25, whereas 1st year scored 5.40 points

```python
plt.bar(df_2.groupby('Year')['score '].mean().index,df_2.groupby('Year')['score '].mean().values)
plt.xticks([1,2,3])
plt.xlabel('Year')
plt.ylabel('Average Score')
plt.title('Average score by year')
plt.tight_layout()
plt.savefig('Average score by year')
```
